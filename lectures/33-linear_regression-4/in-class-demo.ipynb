```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(10, 10)
y = np.random.randn(10)

beta, rank, res, sing = np.linalg.lstsq(x, y)
print(beta)
```

```python
rss = ((y - x.dot(beta))**2).sum()
print(rss)
```

```python
tss = (y ** 2).sum()
print(tss)
```

```python
R2 = 1 - rss/tss
print(R2)
```

```python
# fit using part of the data (here 70%)
x = np.random.randn(1000, 1000)
y = np.random.randn(1000)

beta, rank, res, sing = np.linalg.lstsq(x[:700], y[:700])
print(beta)
```

```python
# then test using the other data (here 30%)
# R^2 is not biased upwards! this is GOOD!
rss = ((y[700:] - x[700:].dot(beta))**2).sum()
tss = (y[700:] ** 2).sum()
R2_oos = 1 - rss/tss
print(R2_oos)
```

# Regression Stability

```python
# what happens if x1 and x2 are similar (but with no noise in y)?
x = np.vstack([[1, 1, 0, 1, 0], 
               [1, 0.99, 0, 1, 0]]).T

y = np.array([1, 1, 0, 1, 0])# + 0.01*np.random.randn(5)

beta, res, rank, sing = np.linalg.lstsq(x, y)
print(beta)
```

```python
# what if x1 and x2 are VERY similar, and there is a little noise in y?
x = np.vstack([[1, 1, 0, 1, 0], 
               [1, 0.999, 0, 1, 0]]).T

y = np.array([1, 1, 0, 1, 0]) + 0.01*np.random.randn(5)
print(y)

beta, res, rank, sing = np.linalg.lstsq(x, y)
print(beta)
print(sing ** -1)
```
