```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
rands = np.random.rand(100, 100000)
```

```python
plt.hist(rands.ravel(), 50);
```

```python
rands.shape
```

```python
rand_means = rands.mean(0)
rand_means.shape
```

```python
plt.hist(rand_means, 50);
```

```python
# let's get weird
rands2 = np.random.rand(500,100000) ** 5
```

```python
rands2.size
```

```python
plt.hist(rands2.ravel(), 100);
```

```python
rands2_means = rands2.mean(0)
plt.hist(rands2_means, 100);
```

```python
rand_flips = np.random.rand(500, 100000) > 0.5
flip_means = rand_flips.sum(0)
plt.hist(flip_means, 50);
```

```python
rand_norms = np.random.randn(100000)
plt.hist(rand_norms, 50);
```

```python
rand_norms.mean()
```

```python
rand_norms_plus5 = rand_norms + 5
plt.hist(rand_norms_plus5, 50);
```

```python
rand_norms_mod2 = rand_norms * np.sqrt(5)
plt.hist(rand_norms_mod2, 50);
print(rand_norms_mod2.var())
```

```python
new_rand_norms = np.random.randn(20, 1000000) + 5
```

```python
new_rand_means = new_rand_norms.mean(0)
new_rand_means.mean()
```

```python
new_rand_variances = new_rand_norms.var(0)
plt.hist(new_rand_variances, 100);
```

```python
new_rand_variances.mean()
```

```python
new_sample_variances = new_rand_norms.var(0, ddof=1)
new_sample_variances.mean()
```
