```python
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
# Create clustered data
cluster_separation = 5 # how separated should clusters be?
n_per_cluster = 50 # how many points per cluster?
scale = [1,1] # scale for each dimension

cluster_means = np.random.randn(5,2) * cluster_separation # cluster means

X = np.vstack([np.random.randn(n_per_cluster,2)*scale + cm 
               for cm in cluster_means])
```

```python
X.shape
```

```python
plt.plot(X[:,0], X[:,1], '.')
plt.axis("square")
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.plot(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
         'kx')
plt.axis("square")
```
