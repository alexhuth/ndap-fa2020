# Convolution + Statistics

The distribution of the sum of two random variables is equal to the convolution of their two distributions. So by the central limit theorem, if you convolve anything with itself a bunch of times (which is like taking the sum of a bunch of random variables), the result will look like the gaussian distribution!

```python
num_convolutions = 50
arr = np.random.rand(6)
arr /= arr.sum()

arr_orig = arr.copy()
for _ in range(num_convolutions):
    arr = np.convolve(arr, arr_orig)

plt.plot(arr, '.-')
```
