```python
# create some fun data with different values along the z-axis
new_data = np.zeros((31,100,100))

for sl in range(31):
    new_data[sl] = sl
```

```python
new_volume = cortex.Volume(new_data, 'S1', 'fullhead')
```

```python
cortex.webshow(new_volume)
```

```python
cortex.quickshow(new_volume);
```
