```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
data1 = np.random.randn(100)
data2 = data1 + np.random.randn(100)*0.5
```

```python
plt.plot(data1, '.-')
plt.plot(data2, '.-');
```

```python
plt.plot(data1, data2, '.');
```

```python
np.corrcoef(data1, data2)[0,1]
```

```python
# permuting (randomly shuffling) one dataset should break
# the correlation
perm_data1 = np.random.permutation(data1)

plt.plot(perm_data1, data2, '.');
```

```python
np.corrcoef(perm_data1, data2)[0,1]
```

# Testing correlation signifiance (exact test)

```python
import scipy.stats
```

```python
scipy.stats.pearsonr(data1, data2) # significant!
```

```python
scipy.stats.pearsonr(perm_data1, data2) # not significant!
```
