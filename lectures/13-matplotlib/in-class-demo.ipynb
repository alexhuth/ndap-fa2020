```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# generate some data to plot
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
```

```python
# plot it!
plt.plot(x, y)
```

```python
# simpler version
simple_x = [0, 1, 2, 3, 4, 5]
simple_y = [0, 5, 1, 4, 2, 3]

plt.plot(simple_x, simple_y)
```

```python
# with a format string
plt.plot(simple_x, simple_y, 'o')
```

```python
# different marker styles
# o , . s
```

```python
plt.plot(simple_x, simple_y, '-o')
```

```python
# line styles
# - -- :
plt.plot(simple_x, simple_y, ':o')
```

```python
# colors
# r g b k
plt.plot(simple_x, simple_y, ':ok')
```

```python
# order does not matter
plt.plot(simple_x, simple_y, 'k:o')
```

# imshow / matshow

```python
image_data = np.arange(110).reshape((11,10))
image_data
```

```python
plt.imshow(image_data)
```

```python
plt.imshow(image_data, vmin=40, vmax=49)
```

```python
plt.cm.#<tab>
```

```python
plt.imshow(image_data, vmin=40, vmax=49, cmap=plt.cm.PuRd)
```

```python
grungy_image_data = image_data.copy()
grungy_image_data[3,4] = 1000
grungy_image_data[7,8] = -1000
```

```python
plt.imshow(grungy_image_data)
plt.colorbar()
```

```python
plt.imshow(grungy_image_data, vmin=0, vmax=110)
```

```python
plt.imshow(grungy_image_data, vmin=0, vmax=110)
plt.colorbar()
```

```python
plt.imshow(grungy_image_data, vmin=0, vmax=110, alpha=0.5) # transparent!
```

# matshow

# xlim / ylim

```python
plt.plot(x, y, '.-') # by default, uses whole range
```

```python
plt.plot(x, y, '.-')
plt.xlim(0, np.pi) # get just the first bump
```

```python
plt.plot(x, y, '.-')
plt.xlim(0, np.pi) # get just the first bump
plt.ylim(0, 1)
```
